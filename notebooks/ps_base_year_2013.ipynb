```python
# Create new HDF5 file
# Copy in all tables other than 

new_store_path = './../data/all_semcog_data_08-17-17_start_2035.h5'

with pd.HDFStore(new_store_path, 'w') as store:
    for name in base.keys():
        excluded_tables = ['/buildings', '/parcels', '/households', '/jobs', '/persons']
        if name not in excluded_tables:
            store[name] = base[name]
```

```python
# Copy 2035 tables

with pd.HDFStore(new_store_path, 'a') as new_store:
    add_tables = ['/buildings', '/parcels', '/households', '/jobs', '/persons']
    for name in add_tables:
        name_2035 = '/2035{}'.format(name)
        new_store[name] = run29[name_2035]
```
